```python
#filter1 = df.loc[(df['Rooms'] == 3) & (df['Type'] == 'h') & (df['Price'] <= 500000)]
filter1 = df.loc[(df['Rooms'] == 3) & (df['Type'] == 'h') & (df['Price'] >= 500000)]
filter1
```

```python
# busca por coluna e palavra especifica dentro dessa coluna
#filter2 = df.loc[df['Address'].str.contains('Turner St|Turner Rd')] 
## Ignorar as letras maisculas e minusculas:
import re # Importar o Regular Expressions para que o flags=re.I funcione
filter2 = df.loc[df['Address'].str.contains('Turner St|Turner Rd', flags=re.I)] 
filter2
```

```python
## Using Regular expressions with conditions
# specific street and price < 
filter3 = df.loc[(df['Address'].str.contains('^59', flags=re.I)) & (df['Price'] <= 500000)] 
# para não aparecer números alem do 59 colocamos o ^ para dizer que inicie com 59
filter3

# Posso utilizar o | (ou)
# filter3 = df.loc[(df['Address'].str.contains('^59', flags=re.I)) | (df['Price'] <= 500000)] 
```

```python
## modifying data frame
#Lemos o data frame
df = pd.read_csv("dataSet's/house.csv")
#e primeiro vamos listar um vendedor em SellerG
df.loc[df['SellerG'] == 'Nelson', 'SellerG'] = 'Ronald'
df
```

```python
# Nova aula
import pandas as pd
import re

df = pd.read_csv("dataSet's/house.csv")
```

```python
df.groupby(['SellerG']).mean('Rooms')
```

```python
# primeiro eu agrupo os vendedores, com Sum somo os valores da coluna Price em ordem crescente
#por causa do ascending = False, e com o head coloco apenas os 10 primeiros
df.groupby(['SellerG']).sum('Price').sort_values('Rooms', ascending = False).head(10)
```

```python
#Soma todos os valores que ele consegue somar com o count()
df.groupby(['SellerG']).count().sort_values('Rooms', ascending = False).head(10)
```
