```python
## nova aula

import pandas as pd
dados = pd.read_csv("dataSet's/fifa.csv")
```

```python
# função de localização .loc 
# primeiro coloca a base e a coluna depois o == ao que você quer
dados.loc[dados['Nationality'] == 'Brazil'] # Da para mudar para dados.loc[dados['Nationality'] == 'Spain']
dados.loc[dados['Age'] == 21]
```

```python
# Valores curtos baseado no que está dentro dos ( ) 
dados.sort_values('Age') # posso adicionar o , ascending = false e isso vai inverter a lista do maior para o menor
dados.sort_values(['Name', 'Age']) # Filtra baseado nos 2, primeiro Nome, depois organiza pela Idade
```
