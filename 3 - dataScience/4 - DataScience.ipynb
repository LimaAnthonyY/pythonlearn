```python
# matplotlib
# https://matplotlib.org/stable/tutorials/index.html
# https://matplotlib.org/stable/gallery/index.html
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv("dataSet's\gasPrices.csv")
gas
```

```python
plt.plot(gas["Year"], gas["Australia"])
plt.plot(gas["Year"], gas["Italy"])
plt.plot(gas["Year"], gas["USA"])
plt.show()
```

```python
# Modificando o grafico
plt.title("Valor do combustivel em US$")

plt.plot(gas["Year"], gas["Australia"], label = "Australia") # Label para a legenda
plt.plot(gas["Year"], gas["Italy"], label = "Italy")
plt.plot(gas["Year"], gas["USA"], label = "USA")
plt.plot(gas["Year"], gas["Japan"], label = "Japan")

plt.xticks(gas['Year'][::2]) # xticks para mudar o eixo X :: para mostrar tudo de 2 em 2 anos

plt.legend() # Ativa a legenda
plt.show()
```

```python
# Adicionando as labels e cores
plt.title("Price in US$")

plt.plot(gas["Year"], gas["Australia"], 'r.-' , label = "Australia")
plt.plot(gas["Year"], gas["Italy"], 'y.-' , label = "Italy")
plt.plot(gas["Year"], gas["USA"], 'b-.' , label = "USA")
plt.plot(gas["Year"], gas["Japan"], 'g.-' , label = "Japan")

plt.xticks(gas['Year'][::2])

plt.xlabel("Years")
plt.ylabel("US$")

plt.legend()
plt.show()
```

```python
# Utilizando o for
plt.title("Price in US$")

listCountry = ['Australia', 'USA', 'Italy', 'France', 'Mexico', 'Germany']

for country in gas:
    if country in listCountry:
        plt.plot(gas["Year"], gas[country], label = [country], marker='.')

plt.xticks(gas['Year'][::2])

plt.xlabel("Years")
plt.ylabel("US$")

plt.legend()
plt.show()
```

```python
# Utilizando o salvar
plt.title("Price in US$")

listCountry = ['Australia', 'USA', 'Italy', 'France', 'Mexico', 'Germany']

for country in gas:
    if country in listCountry:
        plt.plot(gas["Year"], gas[country], label = str([country]), marker='.')

plt.xticks(gas['Year'][::2])

plt.xlabel("Years")
plt.ylabel("US$")

plt.savefig('arquivosGerados/Country.png')
plt.savefig('arquivosGerados/Country.pdf')

plt.legend()
plt.show()
```
